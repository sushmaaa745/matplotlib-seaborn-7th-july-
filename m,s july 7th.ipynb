```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Order_ID': ['ialx566343618', 'akqg208421122', 'njpu434582536', 'rjto796129700', 'zguw716275638'],
    'Agent_Age': [37, 34, 23, 38, 32],
    'Agent_Rating': [4.9, 4.5, 4.4, 4.7, 4.6],
    'Store_Latitude': [22.745049, 12.913041, 12.914264, 11.003669, 12.972793],
    'Store_Longitude': [75.892471, 77.683237, 77.678400, 76.976494, 80.249982],
    'Drop_Latitude': [22.765049, 13.043041, 12.924264, 11.053669, 13.012793],
    'Drop_Longitude': [75.912471, 77.813237, 77.688400, 77.026494, 80.289982],
    'Order_Date': ['2022-03-19', '2022-03-25', '2022-03-19', '2022-04-05', '2022-03-26'],
    'Order_Time': ['11:30:00', '19:45:00', '08:30:00', '18:00:00', '13:30:00'],
    'Pickup_Time': ['11:45:00', '19:50:00', '08:45:00', '18:10:00', '13:45:00'],
    'Weather': ['Sunny', 'Stormy', 'Sandstorms', 'Sunny', 'Cloudy'],
    'Traffic': ['High', 'Jam', 'Low', 'Medium', 'High'],
    'Vehicle': ['motorcycle', 'scooter', 'motorcycle', 'motorcycle', 'scooter'],
    'Area': ['Urban', 'Metropolitian', 'Urban', 'Metropolitian', 'Metropolitian'],
    'Delivery_Time': [120, 165, 130, 105, 150],
    'Category': ['Clothing', 'Electronics', 'Sports', 'Cosmetics', 'Toys']
}

df = pd.DataFrame(data)

# Pair plot for specified columns
sns.pairplot(df[['Agent_Age', 'Agent_Rating', 'Store_Latitude', 'Store_Longitude', 'Delivery_Time']], diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vehicle', y='Delivery_Time', data=df, palette='Set3')
plt.title('Delivery Time Distribution by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Delivery Time')
plt.show()
#box plot
```

```python
#count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, palette='Set1')
plt.title('Number of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#histogram matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['Agent_Age'], bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Agent Age')
plt.xlabel('Agent Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#join plot
sns.jointplot(x='Agent_Age', y='Agent_Rating', data=df, kind='scatter', height=8, color='blue')
plt.show()
```

```python
#kde plot (Kernel Density Estimate)
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Agent_Rating'], shade=True, color='green')
plt.title('Density Plot of Agent Rating')
plt.xlabel('Agent Rating')
plt.ylabel('Density')
plt.show()
```
